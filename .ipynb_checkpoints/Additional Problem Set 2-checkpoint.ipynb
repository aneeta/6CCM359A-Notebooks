### 
Using Monte-Carlo method compute the area of the ellipse

<img src="aps2_1.png">

given be the inequality 
$$x^2 + 2 y^2 < 1$$
Compare your numerical result with the exact value $\pi/\sqrt 2$. Check that the discrepancy is within the expected numerical error. Increase number of points used and check that the error decreases.</p>


```python
def monte_carlo(N):
    n = 0
    for i in range(N):
        x = np.random.uniform(-1,1)
        y = np.random.uniform(-1,1)
        if (x**2 + 2*y**2) < 1:
            n += 1
    
    p1 = n/N
    S1 = 4*p1
    error = 4*np.sqrt((p1*(1-p1))/N)
    solution = np.pi/np.sqrt(2)
    error_actual = S1 - solution
    return p1, error_actual
```

```python
ns = [100*i for i in range(1,200)]
y = [monte_carlo(n)[1] for n in ns]
```

```python
plt.plot(ns, y)
plt.show()
```

### 
Do the same as in 1) for the 3d volume of the shape

<img src="aps2_2.png">

given be the system of inequalities
$$
x^2 + y^2 - z^2 < 1\;\;,\;\; z < 2\;\;,\;\;z>-2
$$
Compare with the exact result $28 \pi/3$, and make the analysis of the error and convergence.

```python
def monte_carlo(N):
    n = 0
    for i in range(N):
        x = np.random.uniform(-2,2)
        y = np.random.uniform(-2,2)
        z = np.random.uniform(-2,2)
        if (x**2 + y**2 - z**2 < 1) and (z < 2 and z > -2):
            n += 1
    p1 = n/N
    S1 = 4**3*p1
    error = 4**3*np.sqrt((p1*(1-p1))/N)
    solution = 28*np.pi/3
    error_actual = S1 - solution
    return p1, error_actual
```

```python
ns = [100*i for i in range(1,100)]
y = [monte_carlo(n)[1] for n in ns]
plt.plot(ns, y)
plt.show()
```

Work out explicitly (analytically) the Gauss nodes $x_i$ and weights $A_i$ for $n=0,\;1$ for the integration measure $w(x)=e^{-x}$ and the range of the integration $[0,\infty)$.
You will have to</p>
<ul>
<li>determine the moments $h_n=\int_{0}^\infty w(x)x^n dx$</li>
<li>Solve the system of equations $\sum_{i=0}^n c_i h_{i+k}=-h_{n+1+k}$ for the coefficients of the polynomial $\phi_{n+1}(x)=c_0+x c_1+\dots+c_n x^n+x^{n+1}$</li>
<li>Find all zeros of the polynomials $\phi_{n+1}(x)=0$, this will give you the nodes $x_i$ for the gaussian integration</li>
<li>Find the weights $A_i$ from the equation $h_m=\sum_{i=0}^n A_i x_i^m,\;m=0,\dots,n$</li>
</ul>

$$\int_0^\infty e^{-x}x^n dx=n!$$

1. Moments $h_n$ <br>
$h_0=\int_{0}^\infty e^{-x} dx = 1$ <br>
$h_1=\int_{0}^\infty xe^{-x} dx = 1$
<p>
2. System of equations <br>

<ul>$n=0$:
    <li>$c_0h_0 = -h_{1}$, $c_0 = \frac{-h_{1}}{h_0}=-1$</li>
    <li>$\phi_{1}(x)=c_0+x = -1 + x$</li>
</ul>

<ul>$n=1$:
    <li>$c_0h_0 + c_1h_1 = -h_{2} $, $c_0 + c_1 = -2$</li>
    <li>$c_0h_1 + c_1h_2 = -h_{3}$, $c_0 + 2c_1 = -6$</li>   
    <li>$c_0 = 2$, $c_1 = -4$</li>
    <li>$\phi_{2}(x)=2-4x+x^2$</li>
</ul>
<p>
3. Roots <br>
$\phi_{1}(1) = 0$, <br>
$\phi_{2}(2 \pm \sqrt{2}) = 0$
<p>
4. Weights
<ul>$m=0$:
    <li>$h_0= A_0 = 1$</li>
</ul>

<ul>$m=1$:
    <li>$h_1= A_0x_0+A_1x_1 = 1$</li>
    <li>$A_0(2 + \sqrt{2})+A_1(2 - \sqrt{2}) = 1$</li>
</ul>
$$A_0=\frac{1}{4}(2+\sqrt 2),\;A_1=\frac{1}{4}(2-\sqrt 2)$$

### 
Adopt the code from the lecture 5.1 to work in the setting described in 3) i.e. for $w(x)=e^{-x}$ and $a=0,\;b=\infty$. Check you analytical results for $A_i$ and $x_i$ for the question 3).</p>
<p>Use those nodes and weight to compute the integral $\int_0^\infty e^{-x} dx$. Compare with the exact result $1/2$. See how the error decreases when you increase $n$</p>


```python
def h(k, a, b):
    return integrate.quad(lambda x: np.e**(-x)*x**k, a, b)[0]
```

```python
def nodes_weights(h, n, a, b):
    # get polynomials
    # Hc=d
    d = [-h(n+1+k, a, b) for k in range(n+1)]
    H = [[h(k+i, a, b) for k in range(n+1)] for i in range(n+1)]
    
    # find polynomial coefficients
    c = np.linalg.solve(H,d) 
    c = np.append(c, [1])
    
    #find roots
    x_i = np.roots(c[::-1]).real
    
    #find weights
    #Xa=h
    h = [h(k,a,b) for k in range(n+1)]
    X = [[x_i[i]**k for i in range(n+1)] for k in range(n+1)]

    a = np.linalg.solve(X,h)
    
    return x_i, a
```

```python
def gaussian(f,A,x):
    return sum(A*np.array(list(map(f,x))))
```

```python
nodes, weights = nodes_weights(h, 20, 0, np.inf)
gaussian(lambda x: np.e**(-x), weights, nodes)
```

```python
_ = [nodes_weights(h, i, 0, np.inf) for i in range(1, 13)]
results = [0.5 - gaussian(lambda x: np.e**(-x), weights, nodes) for (nodes, weights) in _]
```

```python
xs = [i for i in range(1, 13)]
plt.figure(figsize=(10,6))
plt.semilogy(xs, results)
plt.plot(xs, results, "ro")

plt.xlim([0, 13])
plt.ylim([10**-13, 1])
plt.xlabel("$n$", fontsize = 16)
plt.ylabel("Error", fontsize = 16)
plt.show()
```
